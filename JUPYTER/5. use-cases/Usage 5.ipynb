# Usage 5- Markers and symbols

```python
import gmaps
# represent marker location in the map.
marker_locations = [
    
     (42.36009,-71.09416),
     (43.12716,-77.63104),

]
markers = gmaps.marker_layer(marker_locations)
m = gmaps.Map()
m.add_layer(markers)
m
```

```python
# Attach a pop-up box to each marker. Clicking on the marker will bring up the info box. 
specific_university = [
{"city": "Boston", "location": (42.36009,-71.09416), "name": "MIT"},
{"city": "New York", "location": (43.12716,-77.63104), "name": "Simon Business school"}]

university_locations = [univ["location"] for univ in specific_university]
info_box_template = """
<dl>
<dt>City</dt><dd>{city}</dd>
<dt>Name</dt><dd>{name}</dd>
</dl>
"""
univ_info = [info_box_template.format(**univ) for univ in specific_university]
marker_layer = gmaps.marker_layer(university_locations, info_box_content=univ_info)
m = gmaps.Map()
m.add_layer(marker_layer)
m
```
