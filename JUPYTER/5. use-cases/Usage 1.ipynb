# Usage 1  - mark layer using starbucks data

```python
import gmaps
import gmaps.datasets
import pandas as pd
gmaps.configure(api_key="AI...")
a = pd.read_csv("starbucks_us.csv")
b= a[(a['Country Subdivision']=='NY') & (a['City']=='Rochester')]
locations = b[['Latitude','Longitude']]
starbucks_layer = gmaps.symbol_layer(locations, fill_color="green", stroke_color="green", scale=5)
m = gmaps.Map()
m.add_layer(starbucks_layer)
m
```
