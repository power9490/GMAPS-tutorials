# Usage 4  -Weighted heat map layer using earthquakes

```python
import gmaps
import gmaps.datasets
gmaps.configure(api_key="AI...")

df = gmaps.datasets.load_dataset_as_df("earthquakes")

m = gmaps.Map()
heatmap_layer = gmaps.heatmap_layer(df[["latitude","longitude"]], weights = df['magnitude'],
                                   max_intensity = 50, point_radius=6.0)

m.add_layer(heatmap_layer)
m
```
