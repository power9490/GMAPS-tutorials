# Usage 2 - Using direction layer (To Niagara Falls) 

```python
import gmaps
import gmaps.datasets
import pandas as pd
gmaps.configure(api_key="AI...")
mylocation = (43.20,-77.62)
niagara = (43.08,-79.07)
m = gmaps.Map()
my2niagara_layer = gmaps.directions_layer(mylocation,niagara)
m.add_layer(my2niagara_layer)
m
```
