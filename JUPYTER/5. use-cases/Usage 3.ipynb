# Usage 3  -Heat map layer using taxi rides

```python
import gmaps
import gmaps.datasets
gmaps.configure(api_key="AI...")

locations = gmaps.datasets.load_dataset("taxi_rides")

m = gmaps.Map()
m.add_layer(gmaps.heatmap_layer(locations))

m
```
